```python
def angle_diff(a, b):
    """This function calculates the distance between two angles."""
    return (a - b) % 360
```

```python
def circular_distance(theta_1, theta_2):
    """This function calculates the smallest angular distance between two angles."""
    diff = (theta_2 - theta_1) % 360
    if diff > 180:
        diff = diff - 360
    return np.abs(diff)

def closest_theta(theta_1, theta_2, target_theta):
    """This function determines which of the two angles is closest to the target angle."""
    dist1 = circular_distance(theta_1, target_theta)
    dist2 = circular_distance(theta_2, target_theta)
    
    if dist1 < dist2:
        return theta_1
    else:
        return theta_2
```

```python
# Memorization strategy

theta_differences = np.arange(0,365,5)
thetas = np.arange(0,365,5)
choices = np.zeros(shape=(len(theta_differences), len(thetas)))
task1_response = []
task2_response = []

theta_1 = 90
stimulus_1 = 180
for i in range(len(theta_differences)):

    theta_2 = theta_1 + theta_differences[i]

    for j in range(len(thetas)):

        curr_theta = thetas[j]
        chosen_theta = closest_theta(theta_1, theta_2, curr_theta)

        if chosen_theta == theta_1:
            your_classification = (angle_diff(curr_theta, theta_1) < 180)
            choices[i, j] = your_classification
        elif chosen_theta == theta_2:
            your_classification = (angle_diff(curr_theta, theta_2) < 180)
            choices[i, j] = your_classification

    chosen_theta = closest_theta(theta_1, theta_2, stimulus_1)
    if chosen_theta == theta_1:
        your_stimulus1_classification = (angle_diff(stimulus_1, theta_1) < 180)
    elif chosen_theta == theta_2:
        your_stimulus1_classification = (angle_diff(stimulus_1, theta_2) < 180)

    true_task1_stimulus1_classification = (angle_diff(stimulus_1, theta_1) < 180)
    task1_response.append(your_stimulus1_classification == true_task1_stimulus1_classification)
        
    true_task2_stimulus1_classification = (angle_diff(stimulus_1, theta_2) < 180)
    task2_response.append(your_stimulus1_classification == true_task2_stimulus1_classification)

fig, axes = plt.subplots(1,3)
cax = axes[0].imshow(choices.T, cmap='viridis', origin='lower')

xticks = np.arange(0, len(theta_differences), 9)
xtick_labels = np.arange(0, 365, 45)
axes[0].xaxis.set_major_locator(FixedLocator(xticks))
axes[0].xaxis.set_major_formatter(FixedFormatter(xtick_labels))

yticks = np.arange(0, len(thetas), 9)
ytick_labels = np.arange(0, 365, 45)
axes[0].yaxis.set_major_locator(FixedLocator(yticks))
axes[0].yaxis.set_major_formatter(FixedFormatter(ytick_labels))
axes[0].set_xlabel(r'$\theta_2 - \theta_1$')
axes[0].set_ylabel(r'$\theta_t$')
axes[0].set_title("Response to " + r'$\theta_t$')

axes[1].plot(task1_response)
axes[1].xaxis.set_major_locator(FixedLocator(xticks))
axes[1].xaxis.set_major_formatter(FixedFormatter(xtick_labels))
axes[1].set_xlabel(r'$\theta_2 - \theta_1$')
axes[1].set_ylabel("Correct")
axes[1].set_ylim([-0.05,1.05])
axes[1].set_title("Task 1 Response")

axes[2].plot(task2_response)
axes[2].xaxis.set_major_locator(FixedLocator(xticks))
axes[2].xaxis.set_major_formatter(FixedFormatter(xtick_labels))
axes[2].set_xlabel(r'$\theta_2 - \theta_1$')
axes[2].set_ylabel("Correct")
axes[2].set_ylim([-0.05,1.05])
axes[2].set_title("Task 2 Response")

fig.set_figheight(5)
fig.set_figwidth(15)
fig.suptitle("Memorization Strategy")
plt.show()
```

```python
# Performance Comparisons

theta_differences = np.arange(0,365,5)
thetas = np.arange(0,365,5)
task1_performances = np.zeros(shape=(len(theta_differences), 4))
task2_performances = np.zeros(shape=(len(theta_differences), 4))

theta_1 = 90
for i in range(len(theta_differences)):

    true_task1_classification = (angle_diff(thetas, theta_1) < 180)
    true_task2_classification = (angle_diff(thetas, theta_1 + theta_differences[i]) < 180)
    
    # Lazy Strategy
    your_classification = (angle_diff(thetas, theta_1) < 180)
    task1_performances[i,0] = np.sum(your_classification == true_task1_classification)/len(your_classification)
    task2_performances[i,0] = np.sum(your_classification == true_task2_classification)/len(your_classification)

    # Integration Strategy
    integrated_theta = statistics.mean([theta_1, theta_1 + theta_differences[i]])
    your_classification = (angle_diff(thetas, integrated_theta) < 180)
    task1_performances[i,1] = np.sum(your_classification == true_task1_classification)/len(your_classification)
    task2_performances[i,1] = np.sum(your_classification == true_task2_classification)/len(your_classification)

    # Forgetting Strategy
    your_classification = (angle_diff(thetas, theta_1 + theta_differences[i]) < 180)
    task1_performances[i,2] = np.sum(your_classification == true_task1_classification)/len(your_classification)
    task2_performances[i,2] = np.sum(your_classification == true_task2_classification)/len(your_classification)

    # Memorization Strategy
    your_classification = np.zeros(shape=(len(thetas)))
    for j in range(len(thetas)):
        theta_2 = theta_1 + theta_differences[i]
        curr_theta = thetas[j]
        chosen_theta = closest_theta(theta_1, theta_2, curr_theta)

        if chosen_theta == theta_1:
            curr_classification = (angle_diff(curr_theta, theta_1) < 180)
            your_classification[j] = curr_classification
        elif chosen_theta == theta_2:
            curr_classification = (angle_diff(curr_theta, theta_2) < 180)
            your_classification[j] = curr_classification
    task1_performances[i,3] = np.sum(your_classification == true_task1_classification)/len(your_classification)
    task2_performances[i,3] = np.sum(your_classification == true_task2_classification)/len(your_classification)

fig, axes = plt.subplots(1,2)
labels = ["Lazy", "Integration", "Forgetting", "Memorization"]
task1_plot = axes[0].plot(task1_performances, label=labels)
task2_plot = axes[1].plot(task2_performances, label=labels)

xticks = np.arange(0, len(theta_differences), 9)
xtick_labels = np.arange(0, 365, 45)
axes[0].xaxis.set_major_locator(FixedLocator(xticks))
axes[0].xaxis.set_major_formatter(FixedFormatter(xtick_labels))
axes[0].set_xlabel(r'$\theta_2 - \theta_1$')
axes[0].set_ylabel("Fraction Correct Across All " + r'$\theta$')
axes[0].set_ylim([-0.05,1.05])
axes[0].set_title("Task 1 Performance")
axes[0].legend()

axes[1].xaxis.set_major_locator(FixedLocator(xticks))
axes[1].xaxis.set_major_formatter(FixedFormatter(xtick_labels))
axes[1].set_xlabel(r'$\theta_2 - \theta_1$')
axes[1].set_ylabel("Fraction Correct Across All " + r'$\theta$')
axes[1].set_ylim([-0.05,1.05])
axes[1].set_title("Task 2 Performance")
axes[1].legend()

fig.set_figheight(5)
fig.set_figwidth(15)
```

```python
# Lazy strategy with noise

def LazyStrategywithNoise(theta_1, stimulus_1, perceptual_noise_std, decision_noise_std):
    theta_differences = np.arange(0, 365, 5)
    thetas = np.arange(0, 365, 5)
    choices = np.zeros(shape=(len(theta_differences), len(thetas)))
    task1_response = []
    task2_response = []

    theta_1_internal = theta_1 + np.random.normal(0, decision_noise_std)
    stimulus_1_internal = stimulus_1 + np.random.normal(0, perceptual_noise_std)
    for i in range(len(theta_differences)):
        your_classification = (angle_diff(thetas + np.random.normal(0, perceptual_noise_std, len(thetas)), theta_1_internal) < 180)
        choices[i, :] = your_classification

        your_task1_stimulus1_classification = (angle_diff(stimulus_1_internal, theta_1_internal) < 180)
        true_task1_stimulus1_classification = (angle_diff(stimulus_1, theta_1) < 180)
        task1_response.append(your_task1_stimulus1_classification == true_task1_stimulus1_classification)

        your_task2_stimulus1_classification = (angle_diff(stimulus_1_internal, theta_1_internal) < 180)
        true_task2_stimulus1_classification = (angle_diff(stimulus_1, theta_1 + theta_differences[i]) < 180)
        task2_response.append(your_task2_stimulus1_classification == true_task2_stimulus1_classification)

    fig, axes = plt.subplots(1, 3, gridspec_kw={'width_ratios': [1, 1, 1]})
    cax = axes[0].imshow(choices.T, cmap='viridis', origin='lower')

    xticks = np.arange(0, len(theta_differences), 9)
    xtick_labels = np.arange(0, 365, 45)
    axes[0].xaxis.set_major_locator(FixedLocator(xticks))
    axes[0].xaxis.set_major_formatter(FixedFormatter(xtick_labels))

    yticks = np.arange(0, len(thetas), 9)
    ytick_labels = np.arange(0, 365, 45)
    axes[0].yaxis.set_major_locator(FixedLocator(yticks))
    axes[0].yaxis.set_major_formatter(FixedFormatter(ytick_labels))
    axes[0].set_xlabel(r'$\theta_2 - \theta_1$')
    axes[0].set_ylabel(r'$\theta_t$')
    axes[0].set_title("Response to " + r'$\theta_t$')

    axes[1].plot(task1_response)
    axes[1].xaxis.set_major_locator(FixedLocator(xticks))
    axes[1].xaxis.set_major_formatter(FixedFormatter(xtick_labels))
    axes[1].set_xlabel(r'$\theta_2 - \theta_1$')
    axes[1].set_ylabel("Correct")
    axes[1].set_ylim([-0.05, 1.05])
    axes[1].set_title("Task 1 Response")

    axes[2].plot(task2_response)
    axes[2].xaxis.set_major_locator(FixedLocator(xticks))
    axes[2].xaxis.set_major_formatter(FixedFormatter(xtick_labels))
    axes[2].set_xlabel(r'$\theta_2 - \theta_1$')
    axes[2].set_ylabel("Correct")
    axes[2].set_ylim([-0.05, 1.05])
    axes[2].set_title("Task 2 Response")

    fig.set_figheight(5)
    fig.set_figwidth(15)
    fig.suptitle("Lazy Strategy")
    plt.show()

plot = interact(LazyStrategywithNoise, theta_1=(0,360,1), stimulus_1=(0,360,1), perceptual_noise_std=(0, 20.0, 0.1), decision_noise_std=(0, 20.0, 0.1))
```

```python
# Memorization strategy with noise

def MemorizationStrategywithNoise(theta_1, stimulus_1, perceptual_noise_std, decision_noise_std):
    theta_differences = np.arange(0, 365, 5)
    thetas = np.arange(0, 365, 5)
    choices = np.zeros(shape=(len(theta_differences), len(thetas)))
    task1_response = []
    task2_response = []

    theta_1_internal = theta_1 + np.random.normal(0, decision_noise_std)
    stimulus_1_internal = stimulus_1 + np.random.normal(0, perceptual_noise_std)
    for i in range(len(theta_differences)):

        theta_2 = theta_1 + theta_differences[i]
        theta_2_internal = theta_1_internal + theta_differences[i]
    
        for j in range(len(thetas)):
    
            curr_theta = thetas[j] + np.random.normal(0, perceptual_noise_std)
            chosen_theta = closest_theta(theta_1_internal, theta_2_internal, curr_theta)
    
            if chosen_theta == theta_1_internal:
                your_classification = (angle_diff(curr_theta, theta_1_internal) < 180)
                choices[i, j] = your_classification
            elif chosen_theta == theta_2_internal:
                your_classification = (angle_diff(curr_theta, theta_2_internal) < 180)
                choices[i, j] = your_classification
    
        chosen_theta = closest_theta(theta_1_internal, theta_2_internal, stimulus_1_internal)
        if chosen_theta == theta_1_internal:
            your_stimulus1_classification = (angle_diff(stimulus_1_internal, theta_1_internal) < 180)
        elif chosen_theta == theta_2_internal:
            your_stimulus1_classification = (angle_diff(stimulus_1_internal, theta_2_internal) < 180)
    
        true_task1_stimulus1_classification = (angle_diff(stimulus_1, theta_1) < 180)
        task1_response.append(your_stimulus1_classification == true_task1_stimulus1_classification)
            
        true_task2_stimulus1_classification = (angle_diff(stimulus_1, theta_2) < 180)
        task2_response.append(your_stimulus1_classification == true_task2_stimulus1_classification)

    fig, axes = plt.subplots(1, 3, gridspec_kw={'width_ratios': [1, 1, 1]})
    cax = axes[0].imshow(choices.T, cmap='viridis', origin='lower')

    xticks = np.arange(0, len(theta_differences), 9)
    xtick_labels = np.arange(0, 365, 45)
    axes[0].xaxis.set_major_locator(FixedLocator(xticks))
    axes[0].xaxis.set_major_formatter(FixedFormatter(xtick_labels))

    yticks = np.arange(0, len(thetas), 9)
    ytick_labels = np.arange(0, 365, 45)
    axes[0].yaxis.set_major_locator(FixedLocator(yticks))
    axes[0].yaxis.set_major_formatter(FixedFormatter(ytick_labels))
    axes[0].set_xlabel(r'$\theta_2 - \theta_1$')
    axes[0].set_ylabel(r'$\theta_t$')
    axes[0].set_title("Response to " + r'$\theta_t$')

    axes[1].plot(task1_response)
    axes[1].xaxis.set_major_locator(FixedLocator(xticks))
    axes[1].xaxis.set_major_formatter(FixedFormatter(xtick_labels))
    axes[1].set_xlabel(r'$\theta_2 - \theta_1$')
    axes[1].set_ylabel("Correct")
    axes[1].set_ylim([-0.05, 1.05])
    axes[1].set_title("Task 1 Response")

    axes[2].plot(task2_response)
    axes[2].xaxis.set_major_locator(FixedLocator(xticks))
    axes[2].xaxis.set_major_formatter(FixedFormatter(xtick_labels))
    axes[2].set_xlabel(r'$\theta_2 - \theta_1$')
    axes[2].set_ylabel("Correct")
    axes[2].set_ylim([-0.05, 1.05])
    axes[2].set_title("Task 2 Response")

    fig.set_figheight(5)
    fig.set_figwidth(15)
    fig.suptitle("Memorization Strategy")
    plt.show()

plot = interact(MemorizationStrategywithNoise, theta_1=(0,360,1), stimulus_1=(0,360,1), perceptual_noise_std=(0, 20.0, 0.1), decision_noise_std=(0, 20.0, 0.1))
```

```python
# Performance comparisons with noise

def PerformanceComparisonwithNoise(theta_1, perceptual_noise_std, decision_noise_std):

    theta_differences = np.arange(0,365,5)
    thetas = np.arange(0,365,5)
    task1_performances = np.zeros(shape=(len(thetas), 4))
    task2_performances = np.zeros(shape=(len(thetas), 4))
    
    theta_1_internal = theta_1 + np.random.normal(0, decision_noise_std)
    for i in range(len(theta_differences)):

        thetas_internal = thetas + np.random.normal(0, perceptual_noise_std, len(thetas))
        true_task1_classification = (angle_diff(thetas, theta_1) < 180)
        true_task2_classification = (angle_diff(thetas, theta_1 + theta_differences[i]) < 180)
        
        # Lazy Strategy
        your_classification = (angle_diff(thetas_internal, theta_1_internal) < 180)
        task1_performances[i,0] = np.sum(your_classification == true_task1_classification)/len(your_classification)
        task2_performances[i,0] = np.sum(your_classification == true_task2_classification)/len(your_classification)
    
        # Integration Strategy
        integrated_theta = statistics.mean([theta_1_internal, theta_1_internal + theta_differences[i]])
        your_classification = (angle_diff(thetas_internal, integrated_theta) < 180)
        task1_performances[i,1] = np.sum(your_classification == true_task1_classification)/len(your_classification)
        task2_performances[i,1] = np.sum(your_classification == true_task2_classification)/len(your_classification)
    
        # Forgetting Strategy
        your_classification = (angle_diff(thetas_internal, theta_1_internal + theta_differences[i]) < 180)
        task1_performances[i,2] = np.sum(your_classification == true_task1_classification)/len(your_classification)
        task2_performances[i,2] = np.sum(your_classification == true_task2_classification)/len(your_classification)
    
        # Memorization Strategy
        your_classification = np.zeros(shape=(len(thetas)))
        for j in range(len(thetas)):
            theta_2_internal = theta_1_internal + theta_differences[i]
            curr_theta = thetas_internal[j]
            chosen_theta = closest_theta(theta_1_internal, theta_2_internal, curr_theta)
    
            if chosen_theta == theta_1_internal:
                curr_classification = (angle_diff(curr_theta, theta_1_internal) < 180)
                your_classification[j] = curr_classification
            elif chosen_theta == theta_2_internal:
                curr_classification = (angle_diff(curr_theta, theta_2_internal) < 180)
                your_classification[j] = curr_classification
        task1_performances[i,3] = np.sum(your_classification == true_task1_classification)/len(your_classification)
        task2_performances[i,3] = np.sum(your_classification == true_task2_classification)/len(your_classification)
    
    fig, axes = plt.subplots(1,2)
    labels = ["Lazy", "Integration", "Forgetting", "Memorization"]
    task1_plot = axes[0].plot(task1_performances, label=labels)
    task2_plot = axes[1].plot(task2_performances, label=labels)
    
    xticks = np.arange(0, len(theta_differences), 9)
    xtick_labels = np.arange(0, 365, 45)
    axes[0].xaxis.set_major_locator(FixedLocator(xticks))
    axes[0].xaxis.set_major_formatter(FixedFormatter(xtick_labels))
    axes[0].set_xlabel(r'$\theta_2 - \theta_1$')
    axes[0].set_ylabel("Fraction Correct Across All " + r'$\theta$')
    axes[0].set_ylim([-0.05,1.05])
    axes[0].set_title("Task 1 Performance")
    axes[0].legend()
    
    axes[1].xaxis.set_major_locator(FixedLocator(xticks))
    axes[1].xaxis.set_major_formatter(FixedFormatter(xtick_labels))
    axes[1].set_xlabel(r'$\theta_2 - \theta_1$')
    axes[1].set_ylabel("Fraction Correct Across All " + r'$\theta$')
    axes[1].set_ylim([-0.05,1.05])
    axes[1].set_title("Task 2 Performance")
    axes[1].legend()
    
    fig.set_figheight(5)
    fig.set_figwidth(15)
    plt.show()

plot = interact(PerformanceComparisonwithNoise, theta_1=(0,360,1), perceptual_noise_std=(0, 20.0, 0.1), decision_noise_std=(0, 20.0, 0.1))
```
